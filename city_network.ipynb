```python
import networkx as nx
import matplotlib.pyplot as plt

def create_transportation_system_graph():
    # Create an undirected graph
    G = nx.Graph()

    # Add nodes representing locations in the transportation system
    locations = ["Rithala", "Netaji Subhash Place", "Pratap Nagar", "Kashmere Gate", "Welcome", "Dilshad Garden", 
                "Samaypur Badli", "Azadpur", "Vidhan Sabha", "J", "Rajiv Chowk", "INA", "Hauz Khas Enclave", "HUDA City Center",
                "Majlis Park", "O", "P", "Rajouri Garden", "R", "Yamuna Bank",
                "Botanical Garden", "Lajpat Nagar", "V", "W", "X", "Janakpuri", "Dwarka",
                 "Najafagarh", "A2", "A3", "Kalkaji Mandir", "NHPC Chowk", "Raja Nahar Singh",
                ]
    G.add_nodes_from(locations)

    # Add edges representing connections between locations
    connections = [
        ("Rithala", "Netaji Subhash Place", {'weight': 2.5}), ("Netaji Subhash Place", "Pratap Nagar", {'weight': 2.3}), ("Pratap Nagar", "Kashmere Gate", {'weight':  2.0}), ("Kashmere Gate", "Welcome", {'weight':  2.1}), ("Welcome", "Dilshad Garden", {'weight':  1.6}),
        ("Samaypur Badli", "Azadpur", {'weight': 2.3}), ("Azadpur", "Vidhan Sabha", {'weight': 0.5}), ("Vidhan Sabha", "J", {'weight': 0.9}), ("J", "Rajiv Chowk", {'weight': 1.8}), ("Rajiv Chowk", "INA", {'weight': 1.2}), ("INA", "Hauz Khas Enclave", {'weight': 0.9}), ("Hauz Khas Enclave", "HUDA City Center", {'weight': 4.8}),
        ("Majlis Park", "O", {'weight': 1.0}), ("O", "P", {'weight': 1.6}), ("P", "Rajouri Garden", {'weight': 1.7}), ("Rajouri Garden", "R", {'weight': 3.2}), ("R", "Yamuna Bank", {'weight': 1.1}),
        ("Botanical Garden", "Lajpat Nagar", {'weight': 2.0}), ("Lajpat Nagar", "V", {'weight': 1.8}), ("V", "W", {'weight': 2.3}), ("W", "X", {'weight': 2.1}), ("X", "Janakpuri", {'weight': 1.6}), ("Janakpuri", "Dwarka", {'weight': 1.4}),
        ("Najafagarh", "A2", {'weight': 2.7}), ("A2", "A3", {'weight': 5.5}), ("A3", "Kalkaji Mandir", {'weight': 0.7}), ("Kalkaji Mandir", "NHPC Chowk", {'weight': 3.5}), ("NHPC Chowk", "Raja Nahar Singh", {'weight': 5.1}),        
    ]
    
    G.add_edges_from(connections)

    return G

def draw_graph(graph):
    # Draw the graph

    # explicitly set positions
    pos = {"Rithala": (-0.5, 8.5), 
           "Netaji Subhash Place": (1.3, 6), 
           "Pratap Nagar": (3, 4), 
           "Kashmere Gate": (4.5, 5), 
           "Welcome": (6, 6),
           "Dilshad Garden": (7.5, 7),
           
           "Samaypur Badli": (1.5, 9), 
           "Azadpur": (2.8, 8), 
           "Vidhan Sabha": (3.8, 7), 
           "J": (4.5, 5), 
           "Rajiv Chowk": (4.5, 2), 
           "INA": (4.5, -1),
           "Hauz Khas Enclave": (4.3, -4),
           "HUDA City Center": (1, -5),
           
           "Majlis Park": (3.5, 11),
           "O": (2.8, 8), 
           "P": (1.3, 6), 
           "Rajouri Garden": (1.3, 2), 
           "R": (4.5, 2),
           "Yamuna Bank": (6, 3),

           "Botanical Garden": (7.5, 2),
           "Lajpat Nagar": (6, -1), 
           "V": (4.5, 2), 
           "W": (3, 4), 
           "X": (1.3, 2),
           "Janakpuri": (0, -0.1),
           "Dwarka": (-0.8, -3.5),

           "Najafagarh": (-0.8, 4),
           "A2": (0,-0.1), 
           "A3": (4.3, -4), 
           "Kalkaji Mandir": (6, -4), 
           "NHPC Chowk": (7.5, -0.2),
           "Raja Nahar Singh": (7.3, -7.3)
          }

    egde_list1 = [('Rithala', 'Netaji Subhash Place'), ('Netaji Subhash Place', 'Pratap Nagar'), ('Pratap Nagar', 'Kashmere Gate'), ('Kashmere Gate', 'Welcome'), ('Welcome', 'Dilshad Garden')]
    egde_list2 = [('Samaypur Badli', 'Azadpur'), ('Azadpur', 'Vidhan Sabha'), ('Vidhan Sabha', 'J'), ('J', 'Rajiv Chowk'), ('Rajiv Chowk', 'INA'), ('INA', 'Hauz Khas Enclave'), ('Hauz Khas Enclave', 'HUDA City Center')]
    egde_list3 = [('Majlis Park', 'O'), ('O', 'P'), ('P', 'Rajouri Garden'), ('Rajouri Garden', 'R'), ('R', 'Yamuna Bank')]
    egde_list4 = [('Botanical Garden', 'Lajpat Nagar'), ('Lajpat Nagar', 'V'), ('V', 'W'), ('W', 'X'), ('X', 'Janakpuri'), ('Janakpuri', 'Dwarka')]
    egde_list5 = [('Najafagarh', 'A2'), ('A2', 'A3'), ('A3', 'Kalkaji Mandir'), ('Kalkaji Mandir', 'NHPC Chowk'), ('NHPC Chowk', 'Raja Nahar Singh')]

    edges = [egde_list1, egde_list2, egde_list3, egde_list4, egde_list5]
    # edge_labels = nx.get_edge_attributes(graph, "weight")
    # Generate colors for each path
    path_colors = ['red', 'green', 'blue', 'yellow', 'black']
    # nx.draw(graph, pos, edge_labels)
    edge_weights = nx.get_edge_attributes(graph, "weight")

    # Specify colors for nodes
    node_colors = {'Rithala': 'red', 'Netaji Subhash Place': 'red', 'Pratap Nagar': 'blue', 'Kashmere Gate': 'blue', 'Welcome': 'green', 'Dilshad Garden': 'black'}

    all_nodes = [nodes for nodes in graph.nodes]
    print('All nodes: ', all_nodes)
    print(node_colors[all_nodes[2]])
    node_color=['red', 'red', 'red', 'red', 'red', 'red', 
                'green', 'green', 'green', 'green','green', 'green', 'green', 'green',
                'blue', 'blue', 'blue', 'blue', 'blue', 'blue', 
                'yellow', 'yellow', 'yellow', 'yellow', 'yellow','yellow', 'yellow',
                'black', 'black', 'black','black', 'black', 'black']
    print(node_color)
    plt.figure(figsize=(8, 8))
    nx.draw(graph, pos, with_labels=False, node_size=300, font_size=10, font_color='black', linewidths=1, node_color= node_color)

    for i in range(len(path_colors)):
        nx.draw_networkx_edges(graph, pos, edgelist=edges[i], edge_color=path_colors[i], width=3.0)

    nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_weights)

    station_points = {
        'Rithala': [-0.5, 8.9],
        'Netaji Subhash Place': [1.6, 6],
        'Pratap Nagar': [2.8, 4.8],
        'Kashmere Gate': [4.7, 4.8],
        'Welcome': [5.8, 6.6],
        'Dilshad': [7.5, 7.4],

        'Samaypur Badli': [1.5, 9.4],
        'Azadpur': [3, 8.3],
        'Vidhan Sabha': [3.8, 7.4],
        'Rajiv Chowk': [4.8, 1.8],
        'INA': [4.7, -1.3],
        'Hauz Khas Enclave': [4, -4.7],
        'HUDA City Center': [1,-5.7],

        'Majlis Park': [3.5, 11.4],
        'Rajouri Garden': [1.3, 1.4],
        'Yamuna Bank': [6, 3.4], 

        'Botanical': [7.5, 2.4],
        'Lajpat Nagar': [5.7, -1.7],
        'Janakpuri': [0.3, -0.2],
        'Dwarka': [-0.8, -4.1],
        
        'Najafagarh': [-0.8, 4.4],
        'Kalkaji Mandir': [6, -4.7],
        'NHPC': [7.6, -0.6],
        'Raja Nahar': [7.5, -7.3],
        
    }
    
    # plot the stations:
    for stations, value in station_points.items():
        plt.text(value[0], value[1], stations)

    # legends
    l1, = plt.plot([0], [0], label='Rithala - Dilshad', color='red')
    l2, = plt.plot([0], [0], label='Samaypur - HUDA City', color='green')
    l3, = plt.plot([0], [0], label='Majlis - Yamuna', color='blue')
    l4, = plt.plot([0], [0], label='Dwarka - Botanical', color='yellow')
    l5, = plt.plot([0], [0], label='Najafagarh - Raja Nahar', color='black')
    

    plt.legend(handles = [l1, l2, l3, l4, l5], title= "Delhi Metro Map")
    
    # Save the plot as a PNG file
    plt.savefig('example_plot.png')
    plt.show()

if __name__ == "__main__":
    # Create a transportation system graph
    transportation_graph = create_transportation_system_graph()

    # Draw and display the graph
    draw_graph(transportation_graph)
```
